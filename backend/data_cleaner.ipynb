```python
from datetime import datetime

# Convert DECAY_DATE to datetime
df['DECAY_DATE'] = pd.to_datetime(df['DECAY_DATE'], errors='coerce')

# Get today's date
today = pd.Timestamp(datetime.utcnow().date())

# Keep rows where:
# - DECAY_DATE is NaT (no decay recorded) OR
# - DECAY_DATE is in the future
df_active = df[(df['DECAY_DATE'].isna()) | (df['DECAY_DATE'] > today)].copy()

print("Remaining objects in orbit:", df_active.shape[0])
```

```python
from datetime import datetime

# Ensure DECAY_DATE is parsed as datetime
df['DECAY_DATE'] = pd.to_datetime(df['DECAY_DATE'], errors='coerce')

# Get today's date
today = pd.Timestamp(datetime.utcnow().date())

# Filter rows where DECAY_DATE is in the future
df_future_decay = df[df['DECAY_DATE'] > today]

# Number of objects expected to decay in the future
print("Objects with future DECAY_DATE:", len(df_future_decay))
```

```python
# Count non-null (not NaN) values in DECAY_DATE column
non_nan_decay_dates = df['DECAY_DATE'].notna().sum()

print("Non-NaN DECAY_DATE entries:", non_nan_decay_dates)
```

```python
# Count of NaN values in each column
missing_values = df.isna().sum()

# Filter and show only columns with at least 1 missing value
missing_values = missing_values[missing_values > 0]

# Display
print("Columns with missing values:\n")
print(missing_values.sort_values(ascending=False))
```

```python
rcs_counts = df['RCS_SIZE'].value_counts(dropna=False)

# Extract specific counts
medium_count = rcs_counts.get('MEDIUM', 0)
small_count = rcs_counts.get('SMALL', 0)
large_count = rcs_counts.get('LARGE', 0)
nan_count = df['RCS_SIZE'].isna().sum()

# Print the results
print(f"'MEDIUM': {medium_count}")
print(f"'SMALL' : {small_count}")
print(f"'LARGE' : {large_count}")
print(f"NaN     : {nan_count}")



```

```python
# Step 1: Check value counts
print("RCS_SIZE Value Counts:")
print(df['RCS_SIZE'].value_counts())

```

```python
# # Step 2: Find mode (most frequent category)
rcs_mode = df['RCS_SIZE'].mode()[0]
print(f"\nMode of RCS_SIZE: {rcs_mode}")

# Step 3: Fill missing values with the mode
df['RCS_SIZE'].fillna(rcs_mode, inplace=True)


# Fill remaining fields with 'Unknown' for display purposes
fill_cols = ['OBJECT_ID', 'COUNTRY_CODE', 'LAUNCH_DATE', 'SITE']
df[fill_cols] = df[fill_cols].fillna('Unknown')
```
